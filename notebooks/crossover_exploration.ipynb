```python
import numpy as np
import seaborn as sns
```

```python
# Generate random centroids
centroids = np.array([np.random.uniform(0, 1, 2) for _ in range(10)])
centroids[:5]
```

```python
sns.scatterplot(x=centroids[:, 0], y=centroids[:, 1]);
```

```python
centroids_xo = []
while len(centroids_xo) < len(centroids):
    p1, p2 = centroids[np.random.randint(0, centroids.shape[0], 2)]
    if np.random.uniform(0, 1) < 0.3:
        p1[0], p2[1] = p2[1], p1[0]

    o1, o2 = p1, p2
    centroids_xo.append(o1)
    centroids_xo.append(o2)

centroids_xo = np.array(centroids_xo)
centroids_xo[:5]
```

```python
sns.scatterplot(x=centroids[:, 0], y=centroids[:, 1], alpha=1)
sns.scatterplot(x=centroids_xo[:, 0], y=centroids_xo[:, 1], color='red', alpha=0.5);
```

```python
centroids_mut = []
while len(centroids_mut) < len(centroids):
    p1, p2 = centroids_xo[np.random.randint(0, centroids_xo.shape[0], 2)]
    if np.random.uniform(0, 1) < 0.3:
        p1[0], p2[1] = p2[1], p1[0]

    # Mutation: simply to change 1 coordinate randomly
    o1, o2 = p1, p2

    for i, _ in enumerate(o1):
        if np.random.uniform(0, 1) < 0.05:
            o1[i] = np.random.uniform(0, 1)

    for i, _ in enumerate(o2):
        if np.random.uniform(0, 1) < 0.05:
            o2[i] = np.random.uniform(0, 1)

    centroids_mut.append(o1)
    centroids_mut.append(o2)

centroids_mut = np.array(centroids_mut)
centroids_mut[:5]
```

```python
sns.scatterplot(x=centroids[:, 0], y=centroids[:, 1], alpha=1)
sns.scatterplot(x=centroids_xo[:, 0], y=centroids_xo[:, 1], color='red', alpha=0.5);
sns.scatterplot(x=centroids_mut[:, 0], y=centroids_mut[:, 1], color='green', alpha=0.5);
```
